```python
import random
Model_prediction = (random.randint(1,50))
while True:
    number = int(input())
    
    if number > 50 or number <= 0:
        print("Number Invalid")
        continue
    if number > Model_prediction:
        direction = "Bigger than number"
    elif number < Model_prediction:
        direction = "Smaller than number"    
    difference = abs(Model_prediction - number)
    if difference == 0:
        print("VICTORY !!")
        break
    
    if  difference > 25:
        distance = "Too big"
    elif difference > 15:
        distance = "Getting in range"
    elif difference > 5:
        distance = "Close"
    else:
        distance  = "VERY CLOSE !"
    print(f"{distance}, {direction}")
```
